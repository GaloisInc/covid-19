```python
import plotly.figure_factory as ff

import numpy as np
import pandas as pd

df_sample = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/minoritymajority.csv')
df_sample_r = df_sample[df_sample['STNAME'] == 'Ohio']

values = df_sample_r['TOT_POP'].tolist()
fips = df_sample_r['FIPS'].tolist()

colorscale = [
    'rgb(193, 193, 193)',
    'rgb(239,239,239)',
    'rgb(195, 196, 222)',
    'rgb(144,148,194)',
    'rgb(101,104,168)',
    'rgb(65, 53, 132)'
]

fig = ff.create_choropleth(
    fips=fips, values=values, scope=['OH', 'IN', 'MI', 'KY', 'PA', 'WV'],
    binning_endpoints=[14348, 63983, 134827, 426762, 2081313], colorscale=colorscale,
    county_outline={'color': 'rgb(255,255,255)', 'width': 0.5}, round_legend_values=True,
    legend_title='Population by County', title='Ohio and Nearby States'
)
fig.layout.template = None
fig.show()
```

```python
import re

ohioCountyDF = pd.read_csv('ohio_county_list.csv')
ohioCountyDF.columns = ['CTYNAME', 'geo', 'WikiData_pop']
ohioDF = df_sample_r.join(ohioCountyDF.set_index('CTYNAME'), on='CTYNAME')
ohioDF['Lat'] = ohioDF['geo'].apply(lambda x: float(re.match('Point\(([-\d\.]+)\s+([-\d\.]+)\)', x)[2]))
ohioDF['Lon'] = ohioDF['geo'].apply(lambda x: float(re.match('Point\(([-\d\.]+)\s+([-\d\.]+)\)', x)[1]))

hospitalProfiles = pd.read_csv('hospital_profiles_geocoded.csv')
ohioHospitals = hospitalProfiles[hospitalProfiles['State'].str.contains('OH')][hospitalProfiles['Estimated Ventilators'] > 0.0]

ohioHospitals['lon'] = ohioHospitals['lon'].apply(lambda x: x * -1.0)
ohioDF['Lon'] = ohioDF['Lon'].apply(lambda x: x * -1.0)
ohioDF
```
